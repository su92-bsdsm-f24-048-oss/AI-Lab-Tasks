1. Write code for A* Algorithm 

```python
graph = {
    'A': {'B': 1, 'C': 3},
    'B': {'D': 1, 'E': 5},
    'C': {'F': 2},
    'D': {'G': 3},
    'E': {'G': 1},
    'F': {'G': 5},
    'G': {}
}

heuristic = {
    'A': 6,
    'B': 4,
    'C': 4,
    'D': 2,
    'E': 1,
    'F': 3,
    'G': 0
}

def a_star(start, goal):
    open_list = {start: heuristic[start]}
    parent = {start: None}
    g_cost = {start: 0}

    while open_list:
        current = min(open_list, key=lambda x: g_cost[x] + heuristic[x])
        print(f"Visiting Node: {current}")

        if current == goal:
            path = []
            while current:
                path.append(current)
                current = parent[current]
            return path[::-1]

        del open_list[current]

        for neighbor, cost in graph[current].items():
            new_cost = g_cost[current] + cost
            if neighbor not in g_cost or new_cost < g_cost[neighbor]:
                g_cost[neighbor] = new_cost
                total_cost = new_cost + heuristic[neighbor]
                open_list[neighbor] = total_cost
                parent[neighbor] = current
    return None


goal = input("Enter the goal node: ")
path = a_star('A', goal)

if path:
    print("Path found:", path)
else:
    print("No path found.")
```
