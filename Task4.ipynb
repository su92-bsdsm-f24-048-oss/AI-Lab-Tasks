### 1- Code for LUHN Algorithm 

```python
class LuhnAlgorithm:
    def __init__(self):
        self.number = ""

    def get_input(self):
        self.number = input("Enter a card number: ")

    def check_validity(self):
        digits = [int(x) for x in self.number if x.isdigit()]
        total = 0
        reverse = digits[::-1]

        for i, digit in enumerate(reverse):
            if i % 2 == 1:
                digit *= 2
                if digit > 9:
                    digit -= 9
            total += digit

        if total % 10 == 0:
            print("Valid card number")
        else:
            print("Invalid card number")

    def start(self):
        while True:
            print("\n--- LUHN Algorithm ---")
            print("1) Enter Number")
            print("2) Exit")

            choice = input("Choose option: ")

            if choice == "1":
                self.get_input()
                self.check_validity()

            elif choice == "2":
                print("Exiting Program")
                break

            else:
                print("Invalid choice, try again")


app = LuhnAlgorithm()
app.start()
```

### 2- Write a python program to remove punctuations from the given string? 

```python
class PunctuationRemover:
    def __init__(self):
        self.punctuations = "!()-[]{};:'\"\\,<>./?@#$%^&*_~"

    def remove_punctuations(self, text):
        result = ""
        for char in text:
            if char not in self.punctuations:
                result += char
        print("Cleaned text:", result)

    def start(self):
        while True:
            print("\n--- Remove Punctuations ---")
            print("1) Enter Text")
            print("2) Exit")

            choice = input("Choose option: ")

            if choice == "1":
                text = input("Enter text with punctuations: ")
                self.remove_punctuations(text)

            elif choice == "2":
                print("Exiting Program")
                break

            else:
                print("Invalid choice, try again")


app = PunctuationRemover()
app.start()
```

### 3- Write a python program to sort the sentence in alphabetical order? 

```python
class SentenceSorter:
    def __init__(self):
        self.sentence = ""

    def get_sentence(self):
        self.sentence = input("Enter a sentence: ")

    def sort_sentence(self):
        words = self.sentence.split()
        words.sort()
        print("Sorted sentence:", " ".join(words))

    def start(self):
        while True:
            print("\n--- Sort Sentence Alphabetically ---")
            print("1) Enter Sentence")
            print("2) Exit")

            choice = input("Choose option: ")

            if choice == "1":
                self.get_sentence()
                self.sort_sentence()

            elif choice == "2":
                print("Exiting Program")
                break

            else:
                print("Invalid choice, try again")


app = SentenceSorter()
app.start()
```
