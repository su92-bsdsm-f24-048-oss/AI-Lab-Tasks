### Lab 3 Task: 
- • Model-Based Reflex Agent 
- • This agent not only checks the current temperature but also remembers the previous 
action to avoid turning the heater on or off unnecessarily. 

```python
class ModelBasedReflexAgent:
    def __init__(self, desired_temp):
        self.desired_temp = desired_temp
        self.last_action = None

    def perceive(self, current_temp):
        if current_temp < self.desired_temp - 1:
            action = "Turn Heater ON"
        elif current_temp > self.desired_temp + 1:
            action = "Turn Heater OFF"
        else:
            action = "Do Nothing"

        if action == self.last_action:
            return "No Change Needed"

        self.last_action = action
        return action

if __name__ == "__main__":
    agent = ModelBasedReflexAgent(desired_temp=24)
    readings = [22, 23, 24, 25, 24, 23, 22, 22, 23, 24]

    for temp in readings:
        decision = agent.perceive(temp)
        print(f"Temperature: {temp}°C -> {decision}")
```
