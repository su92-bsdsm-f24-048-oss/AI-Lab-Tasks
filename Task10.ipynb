```python
df =df.replace(0, np.nan)
```

```python
null_values = np.sum(pd.isnull(df))
print("Total Null Values:", null_values)
```

```python
df['Timestamp'] = df['Timestamp'].fillna("Unknown")
df['Choose your gender'] = df['Choose your gender'].fillna("Not specified")
df['Age'] = df['Age'].fillna(df['Age'].median())   # numeric â†’ median

df['What is your course?'] = df['What is your course?'].fillna("Unknown")
df['Your current year of Study'] = df['Your current year of Study'].fillna("Unknown")
df['What is your CGPA?'] = df['What is your CGPA?'].fillna("Unknown")
df['Marital status'] = df['Marital status'].fillna("No")

df['Do you have Depression?'] = df['Do you have Depression?'].fillna("No")
df['Do you have Anxiety?'] = df['Do you have Anxiety?'].fillna("No")
df['Do you have Panic attack?'] = df['Do you have Panic attack?'].fillna("No")

df['Did you seek any specialist for a treatment?'] = df['Did you seek any specialist for a treatment?'].fillna("No")
```

```python
df = df.drop('Marital status', axis=1)
```

```python
# Find all object columns
cat_columns = df.select_dtypes(['object']).columns

# Convert them into integers
df[cat_columns] = df[cat_columns].apply(lambda col: pd.factorize(col)[0])
```
