### Task 5
1. DFS with Stack & Node 
2. Research about "Inorder, Preorder, Postorder" and implement in DFS 

```python
graph_map = {
    'P': ['Q', 'R'],
    'Q': ['S', 'T'],
    'R': ['U'],
    'S': ['V', 'W'],
    'T': [],
    'U': ['X', 'Y'],
    'V': [],
    'W': ['Z'],
    'X': [],
    'Y': ['AA'],
    'Z': [],
    'AA': []
}

def dfs_with_stack(start, goal):
    stack = [start]
    parent = {start: None}

    while stack:
        node = stack.pop()
        print(f"Visiting Node: {node}")

        if node == goal:
            path = []
            while node:
                path.append(node)
                node = parent[node]
            return path[::-1]

        for child in reversed(graph_map[node]):
            if child not in parent:
                parent[child] = node
                stack.append(child)
    return None


goal = input("Enter the goal node: ")
path = dfs_with_stack('P', goal)

if path:
    print("Path found:", path)
else:
    print("No path found.")
```

### Task 6
1. BFS without Queue & without Node 
2. BFS with Queue & Node 

```python
class BFS_Without_Queue:
    def __init__(self):
        self.graph = {
            'A': ['B', 'C'],
            'B': ['D', 'E'],
            'C': ['F'],
            'D': [],
            'E': ['F'],
            'F': []
        }

    def bfs(self, start):
        visited = []
        to_visit = [start]

        while to_visit:
            node = to_visit[0]
            to_visit = to_visit[1:]

            if node not in visited:
                print(node, end=" ")
                visited.append(node)
                to_visit.extend(self.graph[node])


obj = BFS_Without_Queue()
print("--- BFS Without Queue & Node ---")
obj.bfs('A')
```

```python
from collections import deque

class Node:
    def __init__(self, value):
        self.value = value
        self.children = []

class BFS_With_Queue:
    def __init__(self):
        self.nodes = {}
        self.create_graph()

    def create_graph(self):
        for val in ['A', 'B', 'C', 'D', 'E', 'F']:
            self.nodes[val] = Node(val)
        self.nodes['A'].children = [self.nodes['B'], self.nodes['C']]
        self.nodes['B'].children = [self.nodes['D'], self.nodes['E']]
        self.nodes['C'].children = [self.nodes['F']]

    def bfs(self, start):
        visited = []
        queue = deque([self.nodes[start]])

        while queue:
            node = queue.popleft()
            if node.value not in visited:
                print(node.value, end=" ")
                visited.append(node.value)
                for child in node.children:
                    queue.append(child)


obj = BFS_With_Queue()
print("\n--- BFS With Queue & Node ---")
obj.bfs('A')
```
