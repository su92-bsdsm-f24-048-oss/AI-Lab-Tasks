### Mini Project 1

```python
class NumberBuzz:
    def __init__(self, end_point):
        self.end_point = end_point

    def run(self):
        for n in range(1, self.end_point + 1):
            word = ""
            if n % 3 == 0:
                word += "Fizz"
            if n % 5 == 0:
                word += "Buzz"
            print(word if word else n)


if __name__ == "__main__":
    print("Running FizzBuzz till 15")
    fb1 = NumberBuzz(15)
    fb1.run()

    print("-" * 25)

    print("Running FizzBuzz till 30")
    fb2 = NumberBuzz(30)
    fb2.run()
```

### Mini Project 2

```python
class FilmBudgetChecker:
    def __init__(self):
        self.films = [
            ("3 Idiots", 55000000),
            ("Lagaan", 25000000),
            ("Gully Boy", 83000000),
            ("Brahmastra", 410000000),
            ("Pathaan", 250000000),
            ("RRR", 350000000),
            ("Baahubali 2", 300000000)
        ]

    def extend_list(self):
        choice = input("Would you like to add extra films? (yes/no): ").strip().lower()
        if choice != "yes":
            print("Using the existing film list.")
            return

        try:
            count = int(input("How many entries do you want to add?: "))
            for n in range(1, count + 1):
                title = input(f"Enter film title #{n}: ")
                while True:
                    budget_input = input(f"Enter budget for '{title}': ")
                    if budget_input.isdigit():
                        self.films.append((title, int(budget_input)))
                        break
                    else:
                        print("Please enter a valid numeric budget.")
        except ValueError:
            print("Please provide a valid number for how many films to add.")

    def budget_report(self):
        if not self.films:
            print("No films in the list to analyze.")
            return

        total = sum(b for _, b in self.films)
        avg = total / len(self.films)
        print(f"\nAverage film budget: ${avg:,.2f}\n")

        above_avg = [f for f in self.films if f[1] > avg]

        if above_avg:
            print("Films with budget higher than average:")
            for title, amt in above_avg:
                diff = amt - avg
                print(f"- {title} (${amt:,}) → Above by ${diff:,.2f}")
        else:
            print("No films exceed the average budget.")

        print(f"\nNumber of films above average: {len(above_avg)}")


if __name__ == "__main__":
    tool = FilmBudgetChecker()
    tool.extend_list()
    tool.budget_report()
```
