```python
import random

class Hangman:
    def __init__(self):
        self.words = ["apple", "banana", "orange", "grapes", "mango"]
        self.word = random.choice(self.words)
        self.guessed = []
        self.tries = 6

    def display_word(self):
        display = ""
        for letter in self.word:
            if letter in self.guessed:
                display += letter + " "
            else:
                display += "_ "
        print(display.strip())

    def play(self):
        print("Welcome to Hangman!")
        while self.tries > 0:
            self.display_word()
            guess = input("Guess a letter: ").lower()

            if guess in self.guessed:
                print("You already guessed that letter")
                continue

            self.guessed.append(guess)

            if guess not in self.word:
                self.tries -= 1
                print(f"Wrong guess! Tries left: {self.tries}")
            else:
                print("Good guess!")

            if all(letter in self.guessed for letter in self.word):
                print(f"You guessed it! The word was '{self.word}'")
                break

        if self.tries == 0:
            print(f"You lost! The word was '{self.word}'")

if __name__ == "__main__":
    game = Hangman()
    game.play()
```
